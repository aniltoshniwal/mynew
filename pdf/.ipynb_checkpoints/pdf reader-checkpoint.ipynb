```python
import pandas as pd
import tabula 
tabula.convert_into("new.pdf", "output.csv", output_format="csv", guess=False)
df=pd.read_csv('output.csv')
data=[]
f_data=[]
m=0
k=0
try:
    for i in df['Consolidated Account Statement'][m:]:
        for j in i.split():
            if j=="Plan":
                l=k-2 
                temp1=[]
                data.append(temp1)
                for i in df['Consolidated Account Statement']:
                    
                    #print(df['Consolidated Account Statement'][l])
                    temp1.append(df['Consolidated Account Statement'][l])
                    
                    l+=1
                    if df['Consolidated Account Statement'][l].split()[0]=="Closing":
                        #print(df['Consolidated Account Statement'][l])
                        temp1.append(df['Consolidated Account Statement'][l])
                        data.append(temp1)
                        l+=6
                        temp1=[]
                        
        k+=1
except:
    pass
[f_data.append(i) for i in data if i not in f_data]
```

```python
json_final={}
json_final['item']=[]
k=0
for p in range(len(f_data)):
    json_data={}
    for i in range(len(f_data[p])):
        if i==0:
            json_data['id']=p+1
            json_data['Fund']=f_data[p][i]
        if i==1:
            if len(f_data[p][i].split(':')[1].strip().split())>=4:
                a=f_data[p][i].split(':')[1].strip().split()
                a=''.join(a[0:-1])   
                json_data[f_data[p][i].split(':')[0].strip()]=a
            else:
                json_data[f_data[p][i].split(':')[0].strip()]=f_data[p][i].split(':')[1].strip()
                
        if i==2:
            #print(p)
            json_data['sub_fund']=f_data[p][i].split('(')[0].strip()
            json_data[f_data[p][i].split('(')[1].split(':')[0]]=f_data[p][i].split('(')[1].split(':')[1].strip(')')
        if i==3:
            json_data[f_data[p][i].split(':')[0].strip()]=f_data[p][i].split(':')[1].strip()
        try:
            if i==4:
                o=1
                txn=[]
                t=0
                for m in f_data[p][4:]: 
                    if m.split(' ')[0]=='Closing':
                        break
                    json_data['txn'+str(o)]={}
                    json_data['txn'+str(o)]['Date']=m.split(' ')[0].strip()
                    json_data['txn'+str(o)]['Amount']=m.split(' ')[-3].strip()
                    json_data['txn'+str(o)]['units']=m.split(' ')[-2].strip()
                    json_data['txn'+str(o)]['price']=m.split(' ')[-1].strip()
                    if t>=1:
                        ub=float(m.split(' ')[-2].strip())+float(txn[-1])
                        u_b='%.3f'%ub
                        json_data['txn'+str(o)]['UnitBalance']=u_b
                        txn.append(u_b)
                    else:
                        json_data['txn'+str(o)]['UnitBalance']=m.split(' ')[-2].strip()
                        txn.append(m.split(' ')[-2].strip())
                    
                    t+=1
                    o+=1
    

        except:
            pass
    json_final['item'].append(json_data)   
json_final
```

```python
import pandas as pd
import tabula 
tabula.convert_into("new.pdf", "output.csv", output_format="csv", guess=False)
df=pd.read_csv('output.csv')
data=[]
f_data=[]
m=0
k=0
try:
    for i in df['Consolidated Account Statement'][m:]:
        for j in i.split():
            if j=="Plan":
                l=k-2 
                temp1=[]
                data.append(temp1)
                for i in df['Consolidated Account Statement']:
                    
                    #print(df['Consolidated Account Statement'][l])
                    temp1.append(df['Consolidated Account Statement'][l])
                    
                    l+=1
                    if df['Consolidated Account Statement'][l].split()[0]=="Closing":
                        #print(df['Consolidated Account Statement'][l])
                        temp1.append(df['Consolidated Account Statement'][l])
                        data.append(temp1)
                        l+=6
                        temp1=[]
                        
        k+=1
except:
    pass
[f_data.append(i) for i in data if i not in f_data]
json_final={}
json_final['item']=[]
k=0
for p in range(len(f_data)):
    json_data={}
    for i in range(len(f_data[p])):
        if i==0:
            json_data['id']=p+1
            json_data['Fund']=f_data[p][i]
        if i==1:
            if len(f_data[p][i].split(':')[1].strip().split())>=4:
                a=f_data[p][i].split(':')[1].strip().split()
                a=''.join(a[0:-1])   
                json_data[f_data[p][i].split(':')[0].strip()]=a
            else:
                json_data[f_data[p][i].split(':')[0].strip()]=f_data[p][i].split(':')[1].strip()
                
        if i==2:
            #print(p)
            json_data['sub_fund']=f_data[p][i].split('(')[0].strip()
            json_data[f_data[p][i].split('(')[1].split(':')[0]]=f_data[p][i].split('(')[1].split(':')[1].strip(')')
        if i==3:
            json_data[f_data[p][i].split(':')[0].strip()]=f_data[p][i].split(':')[1].strip()
        try:
            if i==4:
                o=1
                txn=[]
                t=0
                for m in f_data[p][4:]: 
                    if m.split(' ')[0]=='Closing':
                        break
                    json_data['txn'+str(o)]={}
                    json_data['txn'+str(o)]['Date']=m.split(' ')[0].strip()
                    json_data['txn'+str(o)]['Amount']=m.split(' ')[-3].strip()
                    json_data['txn'+str(o)]['units']=m.split(' ')[-2].strip()
                    json_data['txn'+str(o)]['price']=m.split(' ')[-1].strip()
                    if t>=1:
                        ub=float(m.split(' ')[-2].strip())+float(txn[-1])
                        u_b='%.3f'%ub
                        json_data['txn'+str(o)]['UnitBalance']=u_b
                        txn.append(u_b)
                    else:
                        json_data['txn'+str(o)]['UnitBalance']=m.split(' ')[-2].strip()
                        txn.append(m.split(' ')[-2].strip())
                    
                    t+=1
                    o+=1
    

        except:
            pass
    json_final['item'].append(json_data)   
json_final
```
